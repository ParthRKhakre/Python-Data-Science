# Streamlit 
> Streamlit is a Python library used to quickly build interactive web apps for data science and machine learning using only Pythonâ€”no HTML, CSS, or JavaScript needed.

>Streamlit is built on top of Python, and internally it uses React (JavaScript) for the frontend and WebSockets for real-time interaction between the browser and Python backend.


```python
# Install streamlit -> pip install streamlit

# importing the streamlit library 
import streamlit as st

# Inorder to run the streamlit app use command
# streamlit run app/name.py
```

# Widget
> A widget is a UI element that lets users interact with a Streamlit app, like buttons, sliders, text inputs, checkboxes, etc.

```python
# st.title() 
# is used to display a large main heading at the top of a Streamlit app.
st.title("DashBoard")
```

```python
# st.header() 
# is used to display a section heading in a Streamlit app, smaller than 
# st.title() but larger than normal text.

st.header("Working with streamlit")
```

```python
# st.subheader() is used to display a subsection heading, smaller than 
# st.header() and used to organize content clearly.
st.subheader("Sub-heading goes here")
```

```python
# st.code() is used to display formatted code blocks in a Streamlit app, with syntax highlighting for better readability.
st.code('''
    #include<iostream>
    using namespace std;
    int main(){
        cout<<"Hello World";
    }    
        ''')
```

```python
# st.latex() is used to display mathematical equations written in LaTeX format in a Streamlit app.
st.latex('''
         x^2 + y^2 = 1   
         ''')
```

```python
# st.markdown() is used to display Markdown-formatted text (bold, italics, links, lists, etc.) in a Streamlit app.
st.markdown("""
         
         ### Nifty 50
         - Oracle
         - .Net
         - Nvidia   
            
            """)
```
