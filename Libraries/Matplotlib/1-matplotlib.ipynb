```python
# from a pandas dataframe
batsman = pd.read_csv('sharma-kohli.csv')
batsman.head(10)
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# plotting multiple plot 
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels title
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.plot(batsman['index'],batsman['V Kohli'])

plt.title('Rohit Sharma vs Virat Kohli Carrier Comparison')
plt.xlabel('Session')
plt.ylabel('Runs Scored')
```

```python
# color(hex) and line(width and style) and market
plt.plot(batsman['index'],batsman['RG Sharma'],color='green')
plt.plot(batsman['index'],batsman['V Kohli'],color='red')

plt.title('Rohit Sharma vs Virat Kohli Carrier Comparison')
plt.xlabel('Session')
plt.ylabel('Runs Scored')
```

```python
# types of linestyle : dash,dotted,dashdot,solid
# line width : sets the width of the line 

plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dotted',linewidth=5)
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='dashdot',linewidth=2)

plt.title('Rohit Sharma vs Virat Kohli Carrier Comparison')
plt.xlabel('Session')
plt.ylabel('Runs Scored')
```
