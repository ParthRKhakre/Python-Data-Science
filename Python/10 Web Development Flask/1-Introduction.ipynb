# Flask 
* is a lightweight python web framework used for building web applications and APIs. 
* It is known as a microframework because it provides only the core essentials(like routing and templating).

# Importing Flask

```python
from flask import Flask

# Import the flask class 
# Flask is a WSGI() based micro web framework 
# It lets python acts as HTTP Server
```

# Creating A Flask Object

```python
app = Flask(__name__)
# Flask() creates a flask application object
# __name__ is built in python variable that tell how a file is being executed
# Flask uses it to locate the application module (find application module,set internal app identity)
```

# Flask Route 
@app.route('/') is a decorater that maps the root URL (/) to a Python function that returns an HTTP response.

```python
@app.route('/')
def index():
    return "<h1>Hey</h1>"

# You can add the HTML and CSS in the functions
```

# app.run(debug = True)
* Starting a Flask Development Server on Host:127.0.0.1
and Port:5000
* Blocks the main thread and starts handling HTTP requests
* Flask blocks the thread because it runs a simple synchronous development server, not a production-grade concurrent server.
* It runs Flask in development mode with auto-reload and full error visibility.

```python
app.run(debug=True)
```

# Returning HTML Pages 
* Flask only looks for HTML templates inside templates/ by default. Hence we need to create templates folder
* What Flask Does 
> Loads the file --> Passes it to Jinja2 template engine --> Renders dynamic content --> Returns final HTML response
* The templates/ folder exists because Flask needs a dedicated place for HTML views, processed by Jinja2, cleanly separated from Python logic.

# Integrating HTML Pages With Flask
* We Use render_template() to do so 

```python
from flask import Flask,render_template

app = Flask(__name__)

@app.route('/')
def index():
    return render_template('login.html')
# Inside the render_template('html') provide the html file
```
