# Tkinter
* is a built in-python library that serves as the standard interface to the Tcl/Tk GUI Toolkit

* Generally we name our main python file as app.py


```python
from tkinter import *
```

```python
class NLPApp:
    def __inti__(self):
        
        # login GUI Loading
        
        self.root = Tk()
        # Tk() creates the main window of tkinter application
        # self.root stores that window as an object variable
        # so it can be used anywhere in the class
        
        self.root.title = 'NLP App'
        self.root.iconbitmap("path")
        self.root.geometry('350 x 500')
        self.root.configure(bg='grey')
        
        self.root.mainloop()
        #It starts the Tkinter event loop.
        # Keeps the window open and responsive.
        # Listens for user actions (clicks, typing, closing window).
```
