# Datatypes

1.Definition:

* Datatype are a classification of data which tell the compiler 
or intrepreter how the programmer intends to use the data

* They determines the type of operation that can be performed 
on the data, the values that the data can take and the amount 
of memory needed to store the data

2.Importance of Data Types in Programming

Explaination:

* Data type ensure that data is stored in an efficient way
* They help in performing correct operations on the data
* Proper use of data types can prevent errors and bugs in the program






```python
# Integer
print(1e308)  # 1 * 10^308 maximum amount of value stored in variable of integer type
print(1e309)  # Infinity are value is outof bound

age=35
type(age)
```

```python
# Decimal / Float
print(9.8)
print(1.7e308) # Maximum value 
print(1.7e309) # Infinity

height=5.11
print(height)
print(type(height))
```

```python
# Text/String
print("Hello")
print("P")
# Note : Python doesnot have character type

name="Krish"
print(name)
print(type(name))
```

```python
# Boolean - used in logic building
print(True)
print(False)

is_student = True
print(is_student)
print(type(is_student))
```

```python
is_true = bool()
is_true
```

```python
# bool() is a inbuilt method use to find the boolean value of an variable
# default bool() returns false value
print(bool([]))
print(bool(90))
# Values like 0, "", [], {}, None, and False → False
# Everything else → True
```
