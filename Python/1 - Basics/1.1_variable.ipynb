```python
# Declaring and Assigning Variables 

age = 32
height = 5.11
name="Parth"
is_student = True

# Printing the variables

print("age : ",age)
print("Height: ",height)
print("Name: ",name)
```

```python
# Naming Conventions
# Variable names should be descriptive
# They must start with letter or an '_' and contain letter,numbers and underscores
# variable names are case sensitive

# valid variable names
first_name = "Parth"
last_name = "Khakre"
```

```python
# Understanding Variable types
# python is dynamically typed,type of the variable is detemined at runtime

age = 23     # int
height = 6.1 # float
name="Parth" # string
is_student = True # boolean

print(type(age))
```

```python
# Type Checking and Conversion
type(height)
```

```python
age=23
print(type(age))

# Type conversion
age_str = str(age) 
# converted the int to string a new variable 
# created when we do type conversion

print(age_str)
print(type(age_str))
```

```python
age="23"
print(type(int(age)))

# numeric string can be converted into a integer type
```

```python
name = "krish"
print(type(int(name))) 
# we cannot convert a alphabetic string to a numeric value
```

```python
height=5.11
type(height)

int(height) 
# integer conversion took the floor value

float(int(height))
# conversion of float to int result in loss of data
```

```python
# Dynamic Typing
# Python allows the type of a variable to change as the program executes
 
var=10 # int
print(var,type(var))

var="Hello" # string
print(var,type(var))

var = 3.14
print(var,type(var))
```

```python
# input
age=input("What is the age")
print(age,type(age))

# string is the universal format in the python
age = int(input("Enter age: "))
print(age,type(age))
```

```python
# Simple Calculator

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

sum = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2

print("Sum:",sum)
print("difference: ",difference)
print("product: ",product)
print("Quotient: ",quotient)
```
