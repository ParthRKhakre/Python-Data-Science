# Variables 
are containers where the data is stored
we store anyvalue in the variable using the = 'Assignment operator'

```python
name = 'Nitish'
print(name)

a = 9.23
b = 9.97

sum = a + b
print(sum)
```

In python whenever we required variable we create it 

* Python supports dynamic typing 
* In general statically typed languages we need to declare 
type of data we want to store in variable
but in python we dont do it that what dynamic typing is 

```python
# Dynamic binding
x = 92
print(x)

x = "sam"
print(x)

# here we put 92 in the 'x' variable then changed it to 'sam'
# in python there is no fixed datatype of a variable it can hold 
# data of anytype anytime that what is dynamic binding
```

```python
# Types to create variables

# 1.Individual
a = 10
b = '9'

# 2.Combined 
a,b,c = 90,'parth','A+'

# 3.Similar
a = b = c = 9;
```

```python
# Python Intrepreter is a program that convert the english code into machine code
# Keywords in python are some reserved words that cannot be used for general purpose

# Identifiers/Literals name given by developer to the variables are know as literals
# Rules to write Identifiers
# 1.you can't start with digit
# 2.you can use uppercase lowercase
# 3.No special symbols can be used except _(Underscore)
# 4.Identifiers cannot be keywords

# Input function
# input() a function used to take input from user 
# input is always a string because string is a universal format

# Explicit type conversion or type casting, 
# requires the programmer to explicitly convert 
# a value from one data type to another using built-in functions. 

# Implicit type conversion also known as coercion, 
# is automatically performed by the Python interpreter 
# when operations involve different data types.
print(9.84 + 10)
print(5.6 + int('4'))  

# Take two numbers as input add them and print it on screen
a = int(input("Enter a :")) # anytype of input is automatically converted to string
b = int(input("Enter b :")) # Implicit type conversion

print("sum = ",a + b) # Explicit conversion

# Python type conversion doesnt changes the original data 
# it create a new data 

# Type conversion is done by python
# Type casting is done by the developer explicitly
```

```python
# Literal are the values stored in the variable
a = 0b1010 # Binary Literals
b = 0x12c  # Hexaadecimal 
c = 0o310  # Octal
d = 100    # Decimal 

print(a,b,c,d)

# Float Literal
float_1 = 9.3
float_2 = 1.3e10
float_3 = 4.2e-3

print(float_1,float_2,float_3)

# complex literal 
x = 9 + 91j
print(x,x.real,x.imag)
```
