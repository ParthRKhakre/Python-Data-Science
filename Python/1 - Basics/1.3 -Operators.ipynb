# Operators
are the mathematical symbol use to perform operation on the Operand(Value)

# Arithematic Operator 
(Addition) +,(Subtract) - , * (Multiplication),
/ (Divison) ,// (Integer Division) ,** (Power),% (modulas)

```python
a = 10
b = 10

print("addition: ",a+b)
print("substraction: ",a-b)
print("Multiplication: ",a*b)
print("Division: ",a/b)
print("Integer Division: ",a//3)
print("Power: ",a**b)
print("Modulas: ",a % 5)
```

# Relational Operators
These operators are used to compare the variables


```python
# Relational operator return a true(1) or false(0) value
print(5 > 4)  # greater than
print(5 < 4)  # less than
print(4 >= 5) # greater than equal to 
print(4 <= 5) # less than equal to
print(4 != 5) # not equal to 
print(4 == 5) # equal to
```

# Logical Operator
are the operator used to perform logical operation in program


```python
# In python the logical operator treat all values as True 
# except 0,None,False 
# Logical Operator return 0 if value is false else return true value

print(2 and 0) # AND operator 
print(0 or 30) # OR operator
print(not 9)   # NOT operator
```

# Bitwise Operator
Special Operator that operate over binary values

```python
# bitwise and operator

# In binary "and"
# 010 = 2
# 011 = 3  we perform and operation
# 010 = 2
print(2 & 3)

# In binary "or"
# 010 = 2
# 011 = 3  perform or operation
# 011 = 3
print(2 | 3)


# bitwise xor 
# 010 = 2
# 011 = 3 xor operation was performed
# 001 = 1
print(2 ^ 3)

# bitwise not
# 3 = 011
# ~3 = 100(-4) here there are 1'ones representing the -ve 
print(~3)

# bitwise left shit 
# 100 = 4
# after two left shifts 001 = 1
print(4 >> 2)

# bitwise right shift
# 100 = 4
# after two right shift 10000 = 16
print(4<<2)
```

# Membership Operators
in/not in - tells whether a thing is present or not inside something it is case sensitive in nature

```python
# Does 'D' exist in "Delhi"
print('D' in 'Delhi')
print('N' not in 'Nagpur')
print(3 in [2,1,21,3])
```


Find the sum of a 3 digit number entered by the user 

```python
number = int(input("Enter the 3 digit number : "))

sum = 0

while number != 0 :
    remainder = number  % 10
    sum += remainder
    number //= 10 # absolute division
    
print(sum)    
```
