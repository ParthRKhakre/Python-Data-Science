# Dictionary 
* is a collection of key values,used to stored data values like a map, which unlike other data types which hold only a single value as an element.

In some languages it is known as map or associative arrays

dict = {'name':'nitish','age':33,'gender':'male'}

Characteristics:
* Mutable
* Indexing has no meaning 
* keys cant be duplicated 
* keys cant be mutable items
