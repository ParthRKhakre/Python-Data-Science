# Dictionary 
* is a collection of key values,used to stored data values like a map, which unlike other data types which hold only a single value as an element.

In some languages it is known as map or associative arrays

dict = {'name':'nitish','age':33,'gender':'male'}

Characteristics:
* Mutable
* Indexing has no meaning 
* keys cant be duplicated 
* keys cant be mutable items

```python
# empty dictionary
dic = {}
print(dic)
print(type(dic))
```

```python
# 1D dictionary 
dic1 = {'name':'nitish','age':21,'gender':'male'}
print(dic1)

# 2D dictionary - JSON format
dic2 = {'name':'nita',
        'age':19,
        'friend':{
            'name':'niti',
            'age':20
            }
        }
print(dic2)
print(type(dic2))

# creating dictionary using dict function
dic3 = dict([('name','nitas'),('age',20)])

# duplicate keys are not allowed in dictionary if aded then last key is assigned
d4 = {
    'name':'nitish',
    'name':'ratan'
}
print(d4)

# mutable items cannot be added as key
d5 = {
    'name':'ratan',
#    [1,2,3]:'std'    unhashable type error
    (1,2,3):'CEO'
    }
print(d5)
```

#  Accessing items 
* we dont have index in dictionary so access using index is not possible 

```python
my_dict = {'name':'Jack','age':26}
# my_dict[0] is not allowed 

# Inorder to access the items use [key]
my_dict['name']
my_dict['age']

# Another option to access items is .get() method
# my_dict.get(key) it returns value of the key we have passed 
name = my_dict.get('name')
print(name)
```

# Adding items(key:value) in dictionary 
* As dictionary are mutable hence we can add

```python
print(d4)
# dict_name['key'] = 'value'
d4[3] = 'Might Delete Later'
d4['gender'] = 'male'
d4['weight'] = 60
print(d4)
```

# Remove key-value pair

```python
d = {
    'name': 'ratan', 
    3: 'Might Delete Later', 
    'gender': 'male', 
    'weight': 60
    }

print(d)

# pop(key) - deletes the value with particular provided key 
d.pop(3)
print(d4)

# popitem - delete the last key:value pair 
d.popitem()
print(d)

# del - you clear delete complete dictionary or an particular item
del d['name']
print(d)

# clear - make the dictionary empty
d.clear()
print(d)
```

```python
# Accessing items in 2D dictionary
student = {
    'name':'larry',
    'age': 20,
    'subject':{
        'Web services': 'A+',
        'Microservice': 'A+'
    }
}

grade = student['subject']['Web services']
print(grade)

# Adding in 2D dictionary
student['subject']['Dev'] = 'A+'
print(student)
```

# Dictionary Operations
* Membership - here it works on key not on value
* Iterations - here also it iterates on keys

```python
print(student)
'name' in student
```

```python
d = {'name':'nitish','gender':'male','age':20}

for i in d:
    print(i) # i iterate on keys 
    
# To itearate on values we use dic_name[key]

print('--------------------------')
print("values of d are as below: ")
for i in d:
    print(i,":",d[i])    
```

# Dictionary Functions

```python
# len - total number of key:value pair values
total_itams = len(d)
print(total_itams)
 
# stored - lexicographically sorting based on key
sorted_dic = sorted(d)
print(sorted_dic)

# min - returns the minimum key value lexicographically
min_val = min(d)
print(min_val)

# max - return the maximum key value lexicographically
max_val = max(d)
print(max_val)
```

```python
# items() - returns a list of tuples with key and values
item = d.items()
print(item)

# keys() - returns a list of keys
key = d.keys()
print(key)

# values() - returns a list of values
value = d.values()
print(value)
```

```python
# Update use to permently update dictionary1 by dictionary2
d1 = {1:2,3:4,4:5}
d2 = {4:7,6:8}

d1.update(d2) # change value of d1 using d2
print(d1)
```

# Dictionary Comprehension
* { key:value for vars in iterable }

```python
# print first 10 numbers and their squares
{i:i**2 for i in range(1,11)}
```

```python
# Convert the km from list to miles
distance = {'delhi':1000,'mumbai':2000,'bangalore':4000}
mile = {key:value*0.62 for (key,value) in distance.items()}
print(mile)
```

```python
# using zip() create new dictionary
days = ['sunday','monday','tuesday','wednesday','thrusday','friday','saturday']
temp_c = [30.5,32.4,31.2,33.4,29,34.8,20]

dic = {key:val for (key,val) in zip(days,temp_c)}
print(dic)
```

```python
# using if condition create a new list having items with no stock as 0
product = {'phone':10,'laptop':0,'charger':5,'tablet':0}

available_items = {key:val for (key,val) in product.items() if val != 0} 
print(available_items)
```

```python
# Nested Comprehension
{i:{j:i*j for j in range(1,11)} for i in range(2,5)}
```
