# Loops

```python
# for loop - loop used to iterate over sequence of numbers
range(5) # range function helps us to generate number between (0,5)
```

```python
for i in range(5):
    print(i)
# last value of range is not included    
```

```python
for i in range(-5,5):
    print(i)

# for i in range(starting,endl+1,step): 
# default value of the step function is 1, you can change it as required   
```

```python
for i in range(0,10,2): # step size is 3 
    print(i)
```

```python
for i in range(6,0,-1):
    print(i)
```

# Iterating over strings

```python
# strings 

str = "Parth Khakre"
for i in str:
    print(i)
```

```python
# while loop
# The while loop continues to execute as long as the condition is True.
# Always make sure that you dont get suck into the infinite loop 

count = 0
while(count<10):
    print(count) 
    count = count + 1 

print("total is: ",count)       
```

# Loop Control Statements

```python
# break 
# the break statement exits the loops prematurely

# break statement
for i in range(10):
    if i == 5:
        break # gets us out of the loop
    print(i)
```

```python
# continue 
# continue statement skips the current iteration and continues with the next.

for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
# pass 
# pass statement is a null operation 
# it does nothing just tell intrepreter to not throw an error.

for i in range(5):
    if i == 3:
        print("number is",i)
        pass
    print(i)
```

# Nested Loops

```python
# Nested loops 
# loop inside a loop

for i in range(3):
    for j in range(4):
        print(f"i:{i},j:{j}")
```

```python
# Prime numbers between 1 and 100
for num in range(1, 101):
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            print(num)
```

# Conclusion:

Loops are powerful constructs in python that allows you to execute a block of code multiple times.By understanding and using for and while loops along with loop control statements like break,continue and pass you can handle a wide range of programming tasks efficiently
