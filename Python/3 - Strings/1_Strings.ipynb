# Strings 
text in programming is known as string.  

In Python specifically, String are a sequence of unicode characters 

# Why UNICODE NOT ASCII
ASCII:
ASCII is a character encoding standard that uses 7 bits to represent 128 characters (English letters, digits, symbols). It supports only basic English text.

Unicode:
Unicode is a universal character encoding system that represents characters from almost all languages (English, Hindi, Chinese, emojis, etc.). It supports millions of characters.

Why Python chose Unicode:
Python chose Unicode to support global languages by default, avoid encoding issues, and make string handling simple and consistent across different regions and platforms.

# Accessing Substrings from a string

```python
# positive indexing - python internally assign 0 index to the first character of the string
s = "Hello world"
print(s[0])
print(s[1])
print(s[2])
# if you try to access any index that not part of string it gives a stack overflow error
```

```python
# Negative indexing - python allows negative indexing it assigns the last index as -1 
print(s[-1])
print(s[-2])

# Benefit - Without knowing you can find the last and second last character
```

```python
# Slicing - technique of extracting portion(subset) of element
s = 'Hello world'
print(s[0:5])

# s[starting:ending+1:step_size]
print(s[2:3])
```

```python
print(s[2:]) # here we are starting from 2 and go no till the string ends
print(s[0:]) # from 0 to end of the string
print(s[:3]) # from 0 to index 2
print(s[0:6:2]) # slicing with step size 2

r = s[::-1] # reversing a string
print(r)

print(s[6:0:-2]) # when you put step size negative starting value must be bigger
```

```python
s = 'Hello world'

# print the world using negative indexing
print(s[-5:])
print(s[-1:-6:-1])
```

# Special Functions Related to strings

```python
s = 'hello world'
# Capitalize/Title/Upper/Lower/Swapcase

s.capitalize() # it capitalize the first character of the string 
```

```python
s.title() # title is capitalize first character after a space
```

```python
s.upper() # converted all the character to uppercase
```

```python
s.lower() # converts all the characters to lower case
```

```python
"hElLo wOrLD".swapcase() # it capitalizes the lowercase and make capital lowercase
```

```python
# count - helps in counting frequency of the substring in a string
s = "my name is nitish my"
s.count("i")
s.count("my")
```

```python
# find() it finds the starting index of the substring
s = "my name is nitish my"
s.find('is')

# if the provided input string is not present then it returns -1
```

```python
# index() - it is similar to find() just the difference is if the provided input is not present then it through an error
s.index("is")
s.index('x') # error is thrown as it is not part of the string
```

```python
# endswith()- method that checks if a string ends with a specified suffix and returns True if it does, otherwise returns False.
a = "Hello How are you"
a.endswith('you')
a.endswith('im')
```

```python
# startswith - method that checks if a string starts with a specified suffix and returns True if it does, otherwise returns False.
a.startswith("Hello")
```

```python
# format() -  is used to format strings by inserting values into placeholders defined by curly braces {}

name = "Larry";
day = "monday";
"Hey {} What's the Plan for {}".format(name,day)
```

```python
# isalnum() - check whether a string contains only lowercase letters and no spaces, special characters, or symbols, the method returns True
"nitish".isalnum()
"nitish12".isalnum()
"nitish%".isalnum()
```

```python
# isalpha() - checks whether the string is only alphabetic no numbers allowed
"nitish".isalpha()    # true
"nitish12".isalpha()  # false
```

```python
# isdigit() - checks whether given string consist of only digit no alphabetic or symbolic character is present
s = "123"
v = "hey123"

s.isdigit()
v.isdigit()
```

```python
# s.isidentifier() - returns true if identifier is valid  
"1Nam".isidentifier()
"Num1".isidentifier()
```

```python
# split(parameter_to_split) - returns list of all the substring in the string based on the parameter its space by default
a = "Hey buddy hows going"
a.split()
print(a.split())

# you can split a string based on some character
b = "Hello World owww What a Owl"
b.split('o') # split based on the character 'o'
```

```python
# join() - it is reverse of split() its take list as input and provide the combine form depending on the paramter
" ".join(['Hey', 'buddy', 'hows', 'going'])
```

```python
# replace() - swaps the substring with passed string parameter
"Hey buddy hows going".replace('buddy',"nitish")
```

```python
# strip() - removes the extra spaces very useful in web developement
'nitish                 '.strip()
```
