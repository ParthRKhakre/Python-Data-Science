```python
# Find the length of a given string without using the len() function

val = input("Enter string:")

cnt = 0

for i in val:
    cnt = cnt + 1
 
print("Length of the string is ",cnt)    
```

```python
# Extract username from a given email
# Example if the email is nitish4@gmail.com
# then the username should be nitish4

email = input("enter the email: ")

user = email.split('@')
username = user[0]
print(username) 
```

```python
# count the frequency of a particular character in a prooviided string
# e.g. 'hello how are you' is the string the frequency of h in this string is 2.

word = input("Enter the sentence")
val = input("Enter the word of which you count for : ")
cnt = 0

for i in word:
    if i == val:
        cnt = cnt + 1

print(cnt)        
```

```python
# write a program which can remove a particular character from a string
word = input("Enter the string")
val = input("Enter the word of which you want to remove : ")
res = ' '

for i in word:
    if i != val:
        res = res + i
    
print(res)        
```

```python
# check whether string is a palindrome 

s = 'malayalam'

flag = True

for i in range(0,len(s)//2):
    if s[i] != s[len(s)-i-1]:
        flag = False
        break
   
if flag == False:
    print("Not A Palindrome")
else :
    print("Its a palindrome")           

```

```python
# Write a praogram to count the number of words in a string without split()

s = input("Enter string: ")
lst = []
temp = ''

for i in s:
    if s[i] != ' ':
        temp = temp + i
    else:
        lst.append(temp)
        
lst.append(temp)
```
