# How objects access attributes

```python
class person:
    
    def __init__(self,name_,country_):
        self.name = name_
        self.country = country_
        
    def greet(self):
        if self.country == 'India':
            print('Namaste',self.name)
        else:
            print('Hello ',self.name)                
```

```python
# How to access attributes
person1 = person('Ram','India')

# Object of a class can access attributes and methods of the class
print(person1.name)
print(person1.country)

person1.greet()
```

# Reference Variable
* Reference variable holds the objects
* We can create objects without refernce variable as well 
* An Object can have multiple reference variables
* Assigning a new reference variable to an existing object does not create a new object

```python
# Object without a reference 
class Person:
    def __init__(self):
        self.name = 'nitish'
        self.gender = 'male'
       
Person() # Object gets created 
```

```python
p = Person() 
# Here we created reference to Person() Object

q = p
# Now p,q both are pointing to the same object(Memory Location)
```

```python
print(id(p))
print(id(q))
```

```python
# if we change value of object using any reference it gets reflected in all other references
print(q.name)
print(p.name)
q.name = 'Ratan'
print(q.name)
print(p.name)
```

# Pass By Reference

```python
class Person:
    def __init__(self,name,gender):
        self.name = name
        self.gender = gender
        

# Outside the Person Class
def greet(person):
    print('Hi my name is',person.name,'and I am a',person.gender)
    p1 = Person('ankit','male')
    return p1
    
# You can pass object as an input to functions    
p = Person('Ratan','Male')    
greet(p)            

# Also functions can return objects
x = greet(x)
print(x.name)
print(x.gender)
```

```python
class Person:
    def __init__(self,name,gender):
        self.name = name
        self.gender = gender
        

# Outside the Person Class
def greet(person):
    print(id(person))
    person.name = 'ankit'
    print(person.name)
      
p = Person('Ratan','Male')
print(id(p))    
greet(p)            
print(p.name)
```

```python
class Atm:
    
    # constructor - special function present inside the class
    # In order to execute the function there is no need to call it explicitly
    # Whenever Object is created Constructor is called
    def __init__(self):
        print(id(self))
        self.pin =''
        self.__balance = 0
        # self.menu()
        
    # Rule: Whenever you want to create a variable create it inside the class constructor
    # while creating variable use self keyword
    # Until and unless you create a object of class the code of class have no meaning  
    
    def menu(self):
        user_input = input('''
              Hi How Can I Help You 
              1.Press 1 to create pin
              2.Press 2 to change pin
              3.Press 3 to check balance
              4.Press 4 to Withdraw
              5.Anything else to exit              
              ''')
        if user_input == '1':
            # create pin
            self.create_pin()
        elif user_input == '2':
            self.change_pin()
        elif user_input == '3':
            # check balance
            self.check_balance()
        elif user_input == '4':
            # Withdrawal
            self.withdrawal();
        else:
            # Exit
            exit()
            
    def create_pin(self):
        user_pin = input("Enter Your Pin:") 
        self.pin = user_pin
        
        user_balance = int(input("Enter Your Balance:"))
        self.__balance = user_balance
        print("PIN Created Successfully!")
        self.menu()
        
    def change_pin(self):
        old_pin = input('Enter Old PIN:')
        if old_pin == self.pin:
            new_pin = input("Enter PIN: ")
            self.pin = new_pin 
            print('PIN Changed Successfully')
            self.menu()
        else:
            print("Invalid Existing Pin")  
            self.menu()  
     
    def check_balance(self):
        pin = input("Enter Your PIN")
        if pin == self.pin:
           print(self.__balance)
           self.menu()
        else:
            print('Enter Correct PIN') 
            self.menu()
            
    def withdrawal(self):
        user_pin = input("Enter PIN:")
        if user_pin == self.pin:
            amount = int(input("Enter the Withdrawal Amount: "))
            if amount <= self.balance:
                self.__balance = self.balance - amount
                print("Withdrawal Successful Current Balance",self.balance)
                self.menu()
            else:
                print("Insufficient Balance")        
        else:
            print("Incorrect PIN!")    
        
```

# Static Keyword


```python
class Employee:
    
    # Static Variable its a common variable shared across all the instance of the class
    # It is declared inside the class but outside the method and constructor
    
    __count = 0
    
    
    def __init__(self):
        # Instance Variable
        self.name = 'Nitish'
        self.gender = 'Male' # Instance Variable are create with object
        Employee.__count += 1  # Static variable are created with class
       
    # Below are the static method that are create for working with static variables
    # We need to use @static decorator 
    # static method are utiltity methods     
    @staticmethod    
    def get_count():
        return Employee.__count
    
    @staticmethod
    def set_count(count):
        Employee.__count = count    
```

```python
p1 = Employee()

# you cannot access these function with the following manner 
# p1.get_count()
# Instead you can access them using className not Object name
Employee.get_count()
Employee.set_count(0)


```

```python
p2 = Employee()
p2.count
```
