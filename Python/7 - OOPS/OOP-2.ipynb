# How objects access attributes

```python
class person:
    
    def __init__(self,name_,country_):
        self.name = name_
        self.country = country_
        
    def greet(self):
        if self.country == 'India':
            print('Namaste',self.name)
        else:
            print('Hello ',self.name)                
```

```python
# How to access attributes
person1 = person('Ram','India')

# Object of a class can access attributes and methods of the class
print(person1.name)
print(person1.country)

person1.greet()
```

```python
# WHAT IF WE TRY TO ACCESS ATTRIBUTES THAT ARE NOT A PART OF CLASS
# person1.gender
```

# Attributes creation from outside of the class

```python
person1.gender = 'Male' # added new attributes into the class
```

# Reference Variable

```python
# Object without a reference 
class Person:
    def __init__(self):
        self.name = 'nitish'
        self.gender = 'male'
       
Person() # Object gets created 
```

```python
p = Person() 
# Here we created reference to Person() Object

q = p
# Now p,q both are pointing to the same object(Memory Location)
```

```python
print(id(p))
print(id(q))
```

```python
# 
print(q.name)
print(p.name)
q.name = 'Ratan'
print(q.name)
print(p.name)
```
