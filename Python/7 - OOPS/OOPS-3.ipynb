# Aggregation(Has-A-Relationship)
* One Class is Owner of the Other Class

>Example 

> Customer has a Adderess

> Restaurant Has A Menu

```python
class Customer:
    
    def __init__(self,name,gender,address):
        self.name = name
        self.gender = gender        
# As Address is a complex entity So We Need to create a class of it
        self.address = address
        

class Address:
    def __init__(self,city,pin,state):
        self.city = city
        self.pin = pin        
        self.state = state
        
 
# Customer Has A Address 
Address1 = Address('Gurgaon',100010,'Haryana')
Cust = Customer('Nitish','Male',Address1)        
```
