# Characterstics of a list 
* Ordered
* Changeable/Mutable
* Heterogeneos 
* Can have duplicates
* are dynamic 
* can be nested
* items can be accessed
* can contain any kind of object in python

```python
# Order matters
L = [1,2,3]
L1 = [1,2,3]
L2 = [3,2,1]

print(L == L1)
print(L1==L2)

#  Changeble/Mutable
print(L)
L.append(4)
print(L)

# List is homogeneous in nature
L= [1,"Hello",True,90.2]
print(L)

# Can have duplicate elements
L =  [1,2,2,2,3,4]
print(L)

# Nested
L = [1,2,3,[1,2,3]]
print(L)
```

# Creating a list

```python
# Empty list - list with no element
L = []
print(L)

# 1D list - vector
L = [1,2,3,4,5]
print(L)

# 2D list - matrix
L = [1,2,3,[4,5]]
print(L)

# 3D list
L = [[[1,2],[3,4]],[[5,6],[7,8]]]
print(L)

# Heterogeneous list
L = ['a',1,True,9.2]
print(L)

# Using Type Conversion using list()
print(list("Hello"))
```

# Accessing items from a List

```python
#  Indexing
L = [1,2,3,[4,5]]

# Positive indexing  - moves from right to left starting from 0
print(L[0])
print(L[3][1]) # third index's and its first index 

#  Negative indexing - moves from left to right starting from -1
print(L[-1])
print(L[-1][1]) 

# Example
# L = [[[1,2],[3,4]],[[5,6],[7,8]]]
# print(L[1][1][0])

# Slicing
L = [1,2,3,4,5]
print(L[1:4])
print(L[0::2])
print(L[:3])
print(L[0:])
print(L[::-1]) # Reversing list

```
