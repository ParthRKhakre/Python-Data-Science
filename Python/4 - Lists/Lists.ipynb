# Characterstics of a list 
* Ordered
* Changeable/Mutable
* Heterogeneos 
* Can have duplicates
* are dynamic 
* can be nested
* items can be accessed
* can contain any kind of object in python

```python
# Order matters
L = [1,2,3]
L1 = [1,2,3]
L2 = [3,2,1]

print(L == L1)
print(L1==L2)

#  Changeble/Mutable
print(L)
L.append(4)
print(L)

# List is homogeneous in nature
L= [1,"Hello",True,90.2]
print(L)

# Can have duplicate elements
L =  [1,2,2,2,3,4]
print(L)

# Nested
L = [1,2,3,[1,2,3]]
print(L)
```

# Creating a list

```python
# Empty list - list with no element
L = []
print(L)

# 1D list - vector
L = [1,2,3,4,5]
print(L)

# 2D list - matrix
L = [1,2,3,[4,5]]
print(L)

# 3D list
L = [[[1,2],[3,4]],[[5,6],[7,8]]]
print(L)

# Heterogeneous list
L = ['a',1,True,9.2]
print(L)

# Using Type Conversion using list()
print(list("Hello"))

# list() is an inbuilt python constructor used to create list 
# It takes an iterable and converts it into a list
# It can convert list,string,tuple,range into list 
```

# Accessing items from a List

```python
#  Indexing
L = [1,2,3,[4,5]]

# Positive indexing  - moves from right to left starting from 0
print(L[0])
print(L[3][1]) # third index's and its first index 

#  Negative indexing - moves from left to right starting from -1
print(L[-1])
print(L[-1][1]) 

# Example
# L = [[[1,2],[3,4]],[[5,6],[7,8]]]
# print(L[1][1][0])

# Slicing
L = [1,2,3,4,5]
print(L[1:4])
print(L[0::2])
print(L[:3])
print(L[0:])
print(L[::-1]) # Reversing list

```

# Adding Items from a List

```python
# append - add one item to the last of list 
L = [1,2,3,4,5,6]
L.append(7)
print(L)

L.append([7,6])
print(L)
```

```python
# extend - add multiple items to the last of list
L = [1,2,3,4]
L.extend([5,6,7,8,9])
print(L)

L.extend('Delhi')
print(L)
```

```python
# insert - add element at a desired position

L = [1,2,3,4]
L.insert(1,9) # list.insert(position,item)
print(L)
```

# Editing items in a list

```python
L = [1,2,3,4,5]

# editing with indexing
L[0] = 10
L[-1] = 20

# editing with slicing
L[1:4] = [20,30,90]
print(L)
```

# Deleting items from a List 

```python
# lists are mutable hence items can be added as well as deleted

L = [1,2,3,4,5]
print(L)

# del - is used to delete list 

# indexing
del L[3]
print(L)

# slicing
del L[1:4]
print(L)

del L # deletes the entire list
```

```python
# remove - you can delete the item with value, deleting element without index

L = [1,2,3,4,5]
L.remove(5)
print(L)
```

```python
# pop - removes the provided index item 
# if not provided the index then last element from the list

L = [12,13,14,15]
print(L)

# Index Provided
L.pop(1)
print(L)

# Index not provided
L.pop()
print(L)
```

```python
# clear - instead of deleting list it deletes element and creates a empty list
A = [1,2,3,4,5]
print(A)

A.clear()
print(A)
```

# Operations on List
* Arithemetic
* Membership
* Loop

```python
# Arithemetic Operation (+,*)
L1 = [1,2,3,4,5]
L2 = [6,7,8,9,10]

# Concatenation/Mergeing
print(L1+L2)

# Using *
print(3*L1) 
```

```python
# Memebership Operator
L = [1,2,3,4,[5,6]]
print(3 in L)
print(4 not in L)
print([5,6] in L)
```

```python
# Loops
for i in L:
    print(i)
```

# Functions in List 

```python
L1 = [1,3,2,6,4,5]
L2 = [[1,2,3],[4,5,6]]
```

```python
# len() - returns the total item present in the list
len(L1)
len(L2)
```

```python
# min() - returns items with minimum value
min(L1)
min(L2)
```

```python
# max() - returns items with maximum value
max(L1)
max(L2)
```

```python
# count() - returns total occurance of the items
L = [2,2,3,1,4,1,2,5]
L.count(2)
```

```python
# index() - returns index of the first occurance of item
L.index(2)
L.index(5)
```

```python
# reverse() - it permanently reverses the list
L = [2,2,3,1,4,1,2,5]
print(L)

L.reverse()
print(L)
```

```python
# sorted() - it returns a ascendingly sorted list and it is a temporary operation
sorted(L)
sorted(L,reverse=True)
```

```python
# sort() - it returns a sorted list of elements and its a permanent operation

L = [2,2,3,1,4,1,2,5]
L.sort()
print(L)
```

```python
# copy() - creates a shallow copy of the list 
L1 = [1,2,1,3,2]
L2 = L1.copy()

print(id(L1),id(L2))
```

# List Comprehension
It provides a concise way of creating lists

newlist = [expression for item in iterable if condition  == True]

Advantages of List Comprehension
* most time efficient and space efficient than loops 
* Require fewer line of code 
* Transforms iterative statements into a formula

```python
# Add 1 to 10 numbers to list
newlist = []

for i in range(1,11):
    newlist.append(i)
    
print(newlist)    
    
```
