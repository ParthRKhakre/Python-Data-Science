# Tensors
> is a mathematical object use to represent data with multiple dimensions

```python
import numpy as np
```

```python
# 0D - Scalar
a = np.array(4)

print('dimension of a: ',a.ndim)
```

```python
# 1D - vector
a = np.array([1,2,3,4])

print('dimension of a: ',a.ndim)

# 'a' is a 1D tensor with 4 dimension
# the dimension of the vector 'a' are 4 
# whether as 'a' is a 1D tensor
```

```python
# 2D - Matrix
a = np.array([[1,2,3],[4,5,6]])
print(a)
print('dimension of a: ',a.ndim)
```
